```python
import cv2
import PIL.Image
import matplotlib.pyplot as plt
from yolo import YOLO
```

```python
# Arguments for using with Tiny-Yolo V3
# args = {'model_path':'model_data/yolov3-tiny.h5', 
#         'anchors_path': 'model_data/tiny_yolo_anchors.txt'}

# Arguments for using with Yolo V3
args = {'model_path':'model_data/yolov3.h5', 
        'anchors_path': 'model_data/yolo_anchors.txt'}

yolo = YOLO(**args)
```

```python
cap = cv2.VideoCapture(0)
```

```python
ret, frame = cap.read()
rgb_image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(rgb_image)
plt.show()
```

```python
img = PIL.Image.fromarray(rgb_image)
# img.save('image.png')
# img = PIL.Image.open('image.png')
```

```python
r_image = yolo.detect_image(img)
r_image.show()
```

```python
cap.release()
yolo.close_session()
```
