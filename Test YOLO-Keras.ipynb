```python
import time
import cv2
import PIL.Image
import matplotlib.pyplot as plt
from yolo import YOLO
```

```python
# Arguments for using with Tiny-Yolo V3
args = {'base_path':'',
        'model_path':'model_data/yolov3-tiny.h5', 
        'anchors_path': 'model_data/tiny_yolo_anchors.txt',
        'classes_path': 'model_data/coco_classes.txt'}

# # Arguments for using with Yolo V3
# args = {'base_path':'',
#         'model_path':'model_data/yolov3.h5', 
#         'anchors_path': 'model_data/yolo_anchors.txt',
#         'classes_path': 'model_data/coco_classes.txt'}


# # Arguments for using with Yolo V3 OpenImages
# args = {'base_path':'',
#         'model_path':'model_data/yolov3-openimages.h5', 
#         'anchors_path': 'model_data/yolo_anchors.txt',
#         'classes_path': 'model_data/openimages.names.txt'}

# # Arguments for using with Yolo V3 SPP
# args = {'base_path':'',
#         'model_path':'model_data/yolov3-spp.h5', 
#         'anchors_path': 'model_data/yolo_anchors.txt',
#         'classes_path': 'model_data/coco_classes.txt'}

yolo = YOLO(**args)
```

```python
cap = cv2.VideoCapture(0)
```

```python
cap.open(0)

ret = 0
while not ret:
    ret, frame = cap.read()
    time.sleep(.05)
    
rgb_image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(rgb_image)
plt.show()
```

```python
img = PIL.Image.fromarray(rgb_image)
# img.save('image.png')
# img = PIL.Image.open('image.png')
```

```python
img = PIL.Image.open('test.png')
```

```python
r_image = yolo.detect_image(img)
```
